```python
#importing necessary libraries
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#sample dataset (User-item ratings)
data = {
    'Item1':[5, 4, 0, 1],
    'Item2':[4, 0, 0, 1],
    'Item3':[1, 1, 0, 5],
    'Item4':[0, 0, 5, 4]
}
```

```python
#create dataframes (rows=users, columns=items)
df = pd.DataFrame(data, index=['User1', 'User2', 'User3', 'User4'])
```

```python
print("User_Item Matrix:\n", df)
```

```python
#compute similarity between users using cosine similarity
user_similarity = cosine_similarity(df)
```

```python
#convert similarity matrix into dataframes
user_similarity_df = pd.DataFrame(user_similarity, index=df.index, columns=df.index)
```

```python
print('\nUser Similarity Matrix:\n', user_similarity_df)
```

```python
#function to recommend items for a user
def recommend(user):
  similar_users = user_similarity_df[user].sort_values(ascending=False)

  similar_users = similar_users.drop(user)

  most_similar_users = similar_users.index[0]

  print(f"\nMost similar users to {user} is {most_similar_users}")

  #recommended items liked by similer user but not by current user
  recommendations = []
  for item in df.columns:
    if df.loc[user, item] == 0 and df.loc[most_similar_users, item] > 0:
      recommendations.append(item)

    return recommendations
```

```python
user = 'User3'
recommended_items = recommend(user)
print(f"\nRecommended items for {user}: {recommended_items}")
```
